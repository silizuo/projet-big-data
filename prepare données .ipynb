```python
import pickle
import pandas as pd 
import numpy as np 
# Shared folder to read data
SHARED_FOLDER = './yelp'

# Your folder to write new data and create notebooks.
MY_FOLDER = './output'
```

# add a single category to business

```python
with open ('{}/more_business.pickle'.format(SHARED_FOLDER), 'rb') as f:
    business_df = pickle.load(f)
```

```python
business_df
```

```python
with open ('{}/categories.pickle'.format(SHARED_FOLDER), 'rb') as f:
     categories_df = pickle.load(f)
```

```python
business_categories_df = business_df.merge(categories_df, on='business_id', how='left',
                                         suffixes=['business','categories'])
business_categories_df
```

```python
business_categories_df.to_csv('./output/business_categories_df.csv')
```

# category en 5 type

```python
def five_cate(x):
    if (x=='Fast Food'or x=='Pizza'or x=='Chicken Wings'or x=='Sandwiches'or x=='Burgers'or x=='Hot Dogs'):
        x='Fast Food'
    if (x=='Vegetarian'or x=='Vegan'):
        x='Vegeterian'
    if (x=='Coffee & Tea'or x=='Cafes'):
        x='coffee'
    if (x=='Mexican'or x=='Tex-Mex'):
        x='mexican'
    if (x=='Nighlife'or x=='Bars'or x=='Gastropubs'or x=='Dive Bars'or x=='Wine Bars'or x=='Hot Dogs'):
        x='bars'
    return(x)
```

```python
business_categories_df['category'].apply(five_cate)
```

```python
business_categories_df['category1']=business_categories_df['category'].apply(five_cate)
```

```python
business_categories_df
```

```python
business_categories_df.to_csv('./output/business_new_categories_df.csv')
```

# fast food->bon et mauvias

```python
category_fastfood=business_categories_df[business_categories_df['category1']=='Fast Food']
category_fastfood
```

```python
mean=np.mean(category_fastfood.stars)
def label_df(x):
    if x<=mean:x=0
    elif x>mean:x=1
    return(x)
```

```python
category_fastfood['labels']=category_fastfood['stars'].apply(label_df)
```

```python
bon_category_fastfood=category_fastfood[category_fastfood['labels']==1]
bon_category_fastfood.to_csv('./output/bon_category_fastfood.csv')
```

```python
mauvais_category_fastfood=category_fastfood[category_fastfood['labels']==0]
mauvais_category_fastfood.to_csv('./output/mauvais_category_fastfood.csv')
```
