```python
import pickle
import pandas as pd 
import numpy as np 
# Shared folder to read data
SHARED_FOLDER = './yelp'

# Your folder to write new data and create notebooks.
MY_FOLDER = './output'
```

```python
#with open ('{}/business_review_tip_checkin_df.pickle'.format(SHARED_FOLDER), 'rb') as f:
    #business_df = pickle.load(f)
business_df=pd.read_csv('{}/business_df.csv'.format(SHARED_FOLDER))
business_df
```

```python
business_df['labels']=0
```

```python
def label_df(x):
    if x<=2:x=0
    elif (x>2 and x<=3):x=1
    else:x=2
    return(x)
```

```python
business_df['labels']=business_df['stars'].apply(label_df)
```

```python
bon_restaurant=business_df[business_df['labels']==2]
bon_restaurant
```

```python
bon_restaurant.to_csv('./output/bon_restaurant.csv')
```
